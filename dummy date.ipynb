## 【問題1】乱数の作成

```python
import numpy as np
ave = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
n_sample = 500

date0 = np.random.multivariate_normal(ave, cov, n_sample)
print(date0.shape)


#シード値
np.random.seed(0)
print(date0)
```

## 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。


散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(date0[:,0], date0[:,1])
plt.show()
```

## 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。


ヒストグラムはplt.hist()を使うことで描けます。


matplotlib.pyplot.hist — Matplotlib 3.1.1 documentation


各次元ごとに作成し、2つのヒストグラムとしてください。


それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.figure(figsize=(6, 4))
plt.hist(date0[: ,0],bins=50)
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x1")
plt.xlim(-5.5, 2.5)
plt.ylim(0, 30)
plt.show()

#plt.figure(figsize=(6, 4))
plt.hist(date0[: ,1],bins=50)
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")
plt.xlim(-5.5, 2.5)
plt.ylim(0, 32)
plt.show()
```

## 【問題4】データの追加

たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。


$$\begin{bmatrix}
1.0 & 0.8 \\ 0.8 & 1.0
\end{bmatrix}$$

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。

```python
import numpy as np

ave = (0, -3)
cov = [[1.0, 0.8], [0.8, 1.0]]
n_sample = 500

date1 = np.random.multivariate_normal(ave, cov, n_sample)
print(date1.shape)

#シード値
np.random.seed(0)
print(date1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ave0 = (-3, 0) #1問目
ave1 = (0, -3) #4問目 
cov = [[1.0, 0.8], [0.8, 1.0]]
n_sample0 = 500 #1
n_sample1 = 500

date0 = np.random.multivariate_normal(ave0, cov, n_sample0)
date1 = np.random.multivariate_normal(ave1, cov, n_sample1)
np.random.seed(0)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(date0[:,0], date0[:,1], label=0)
plt.scatter(date1[:,0], date1[:,1], label=1)
plt.legend()
plt.show()
```

## 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ave0 = (-3, 0) #1問目
ave1 = (0, -3) #4問目 
cov = [[1.0, 0.8], [0.8, 1.0]]
n_sample0 = 500 #1問目
n_sample1 = 500 #4問目

date0 = np.random.multivariate_normal(ave0, cov, n_sample0)
date1 = np.random.multivariate_normal(ave1, cov, n_sample1)
np.random.seed(0)

join_0 = np.random.multivariate_normal(ave0, cov, n_sample0)
join_1 = np.random.multivariate_normal(ave1, cov, n_sample1)

join_mix = np.concatenate([join_0, join_1])
print(join_mix.shape)
print(join_mix)
```

## 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。


機械学習に使用するデータセットはこのような形になっていることが多いです。


《作成例》


データが3個+3個の場合は以下のような(6, 3)のndarrayになります。

```python
#問題１のラベル付け :0
label_0 = np.insert(join_0, 2, 0, axis=1)
print("問題１の配列数の確認:{}".format(label_0.shape))
print(label_0)
print()

#問題４のラベル付け :1
label_1 = np.insert(join_1, 2, 1, axis=1)
print("問題4の配列数の確認:{}".format(label_1.shape))
print(label_1)


#問題1と4の複合ラベル付け 
label_mix = np.concatenate([label_0, label_1])

print("問題1と4の配列数の確認:{}".format(label_mix.shape))
print(label_mix)
```
