# 折紙問題
## 【問題1】べき乗の算術演算子を使用した実装
### ・1枚の紙を43回折り曲げた時の厚さを計算するプログラム 

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 ** 43
print("厚さ： {}メートル".format(folded_thickness))
```

## 【問題2】単位の変換
### ・キロメートル に変換したコード

```python
print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
```

## 【問題3】for文を使用して作成

```python
THICKNESS = 0.00008
for _ in range(43):
    THICKNESS *= 2
print("厚さ: {}メートル".format(THICKNESS))
```

## 【問題4】計算時間の比較
### ・実行速度を比較

```python
import time
start = time.time()
#####
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 ** 43
print("厚さ： {}メートル".format(folded_thickness))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

## 【問題5】リストへの保存
### ・過程の値合計44個を記録するコード

```python
THICKNESS_list = []

THICKNESS = 0.00008
THICKNESS_list.append(THICKNESS)
for _ in range(43):
    THICKNESS *= 2
    
    THICKNESS_list.append(THICKNESS)
print(THICKNESS_list)

print(len(THICKNESS_list), str("個"))
```

## 【問題6】折れ線グラフの表示
### ・厚さがどのように変化しているか　Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(THICKNESS_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

## 【問題7】グラフのカスタマイズ
### ・3種類

```python
# 赤い折れ線グラフを表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(THICKNESS_list, color='red')
plt.show()
```

```python
#軸の値を大きく表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20) #軸の値に関する設定
plt.plot(THICKNESS_list, color='blue')
plt.show()
```

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(direction='out', length=6, width=2, colors='r',
               grid_color='r', grid_alpha=0.5)
plt.plot(THICKNESS_list, color='purple')
plt.show()
```
